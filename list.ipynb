<h1>List</h1>
<p>The most basic data structure in Python is the sequence. Each element of a sequence is assigned a number - its position or index. The first index is zero, the second index is one, and so forth. Python has six built-in types of sequences, but the most common ones are lists and tuples. There are certain things you can do with all sequence types. These operations include indexing, slicing, adding, multiplying, and checking for membership. In addition, Python has built-in functions for finding the length of a sequence and for finding its largest and smallest elements.</p>
<p>The list is a most versatile datatype available in Python which can be written as a list of comma-separated values (items) between square brackets. Important thing about a list is that items in a list need not be of the same type. Creating a list is as simple as putting different comma-separated values between square brackets.</p>

```python
list1 = [1, 2, 3, 4, 5]
list2 = ["RCB", "CSK", "MI", "KKR", "SRH"]
list3 = [1, "RCB", 2, "CSK", 3, "MI", 4, "KKR", 5, "SRH"]
```

<p>Similar to string indices, list indices start at 0, and lists can be sliced, concatenated and so on.</p>
<h3>Accessing Values in Lists</h3>
<p>To access values in lists, use the square brackets for slicing along with the index or indices to obtain value available at that index. For example:</p>

```python
print(list1[0])
print(list2[0])
print(list3[0])
```

<h3>Slicing the list</h3>
<p>"[:]" is used for slicing. There are 2 positions in the square brackets with ":" as reference. <ul><li>Before ":"</li><li>After ":"</li></ul> The Before positions tells us the starting point of slicing and the after position tell us about the ending index excluding the index. For example</p>
<p>If the Before position is kept blank it indicates the starting index is 0</p>
<h3>Extended Slicing</h3>
<p>It extracts a specific portion of a list based on a set stride or step.</p><p>In Python, the [::] syntax is used for extended slicing.</p>
<p> <b>The Slicing Structure</b> </p> <p> The full syntax inside the brackets is [start:stop:step]: <ul>
<li>start: The index where the slice begins (inclusive)</li>
<li>stop: The index where the slice ends (exclusive).</li>
<li>step: The increment value determining how many items to skip.</li>
</ul> <br>When you omit the values but keep the colons as [::], Python automatically fills in the defaults:
<ul>
    <li>start defaults to the beginning of the list (0).</li>
    <li>stop defaults to the end of the list.</li>
    <li>step defaults to 1</li>
</ul>
</p>
<p> The Extended slicing method is very useful if you want to reverse a list as fast as you can.</p>
<p>Setting the step value to -1 tells Python to read the list from right to left. This is one of the fastest and most common ways to reverse an iterable in Python.</p>

```python
print(list1[0:4])
# Prints elements from index 0 to index 3 excluding index 4
print(list2[:3])
# Prints elements from index 0 to index 2 excluding index 3
print(list3[2:10:3])
#Extended slicing with a step of 3 starts from index 2 and goes up to index 10 excluding index 10
print(list3[::3])
#Extended slicing with a step of 3 starts from index 0 and goes up to last index excluding last index
reversed_list = list3[::-1]  
print(reversed_list)
# Slices backward with a step of 1

```

<h3>Updating Lists</h3>
<p>You can update single or multiple elements of lists by giving the slice on the left-hand side of the assignment operator. For example, Assume you have a list of Names and You want to REPLACE a name called "QWERTY" with "FOX"</p>
<p><b>My approach</b></p>
<p><b>Step 1 :</b> Find the Index of the name which is present in the list </p>
<p><b>Step 2 :</b> Access the element in the list by using it's index and assign it with new "LITERAL"</P>
<p><b>Getting the index of any element if it is present in the list</b></p>
<p> Use <i>list.index(obj)</i> function to get the index of the required object </p>
<p> The method index() returns the lowest index in list that obj appears.</p>

```python
names = ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Heidi", "Ivan", "Judy", "QWERTY"]
print("List before updation:", names)
index = names.index("QWERTY")
print("Index of 'QWERTY':", index)
names[index] = "FOX"
print("List after updation:", names)
```

<h3>Addition of new objects/literals</h3>
<p>We can add any element to a list. But only at the end</p>
<p>We can't add objects in the middle of the list</p>
<p>The addition is done by using <b><i>.append(obj)</i></b> method</p>
<p>It adds the object to the list but at the very end</p>

```python
names = ["Alice", "Bob", "Charlie"]
print("List:", names)
names.append("David")
print("List after appending 'David':", names)
```

<h3>Deleting List Elements</h3>
<p>To remove a list element, you can use either the del statement if you know exactly which element(s) you are deleting or the remove() method if you do not know.</p>
<p>Ways to delete elements in a list<ul>
<li><b>del list[index]</b> - It is used to delete an element in the list if you know the index of the element.</li>
<li><b>.pop(index)</b> - If you know the index of the element that needs to be deleted then even without using del statement ues can use "list.pop(index)" which just deleted the indexed element in the list.</li>
<li><b>.remove(element)</b> - for both pop and del method you need to know the index of the element but for remove methon you just need to know which element or object you need to delete from the list</li>
</ul></p>

```python
fruits = ["Apple", "Banana", "Cherry"]
print("List:", fruits)
#Appending a fruit to the list
fruits.append("Mango")
print("List after appending 'Mango':", fruits)
#Removing a fruit from the list using the remove() method
fruits.remove("Banana")
print("List after removing 'Banana':", fruits)
#Removing a fruit from the list using the pop() method
fruits.pop(1)
print("List after popping element at index 1:", fruits)
#Removing a fruit from the list using the del statement by getting the index of the element to be deleted
index_of_cherry = fruits.index("Apple")
del fruits[index_of_cherry]
print("List after deleting 'Apple':", fruits)
```

<h3>Basic List Operations</h3>
<p>Lists respond to the + and * operators much like strings; they mean concatenation and repetition here too, except that the result is a new list, not a string.</p>
<p>Some of them are :</p>
<p><ul>
    <li>Length : len(list) : Used to find out the length of the list.</li>
    <li>Concatenation : list3 = list1 + list2 : Adds the elements of list2 to the list 1 and stores in in another list</li>
    <li>Repetition : list1*3 : Repeats all the elements in the list for 3 times.</li>
    <li>Membership : i in list --> True : Checks wheather the element is in list or not</li>
    <li>Iteration : for i in list: print i : Allows us to perform loops and operations </li>
</ul></p>

```python
sports = ["Football", "Basketball", "Tennis", "Cricket"]
print("List:", sports)
#Basic operations on the list
# Length of the list
print("Length of the list:", len(sports))
# Concatenation of two lists
more_sports = ["Hockey", "Baseball"]
all_sports = sports + more_sports
print("Concatenated list:", all_sports)
# Repetition of the list
repeated_sports = sports * 2
print("Repeated list:", repeated_sports)
# Membership testing
print("Is 'Tennis' in the list?", "Tennis" in sports)
# Iterating through the list
for sport in sports:
    print(sport)
```

<h3>List methods</h3>
<ol>
    <li>Appending</li>
        <ul>
            <li>Syntax : List.append(obj)</li>
            <li>Description : The method append() appends/adds a passed obj into the existing list.</li>
            <li>Parameters : obj -- This is the object to be appended in the list. </li>
            <li>Return value :This method does not return any value but updates existing list.</li>
        </ul>
    <li>Count</li>
        <ul>
            <li>Syntax : list.count(obj)</li>
            <li>Description : The method count() returns count of how many times obj occurs in list.</li>
            <li>Parameters : obj -- This is the object to be counted in the list.</li>
            <li>Return value : This method returns count of how many times obj occurs in list.</li>
        </ul>
    <li>Extend</li>
        <ul>
            <li>Syntax : list.extend(seq)</li>
            <li>Description : The method extend() appends the contents of seq to list.</li>
            <li>Parameters : seq -- This is the list of elements</li>
            <li>Return Value : This method does not return any value but add the content to existing list.</li>
        </ul>
    <li>Index</li>
        <ul>
            <li>Syntax : list.index(obj)</li>
            <li>Description : The method index() returns the lowest index in list that obj appears.</li>
            <li>Parameters : obj -- This is the object to be find out.</li>
            <li>Return value : This method returns index of the found object otherwise raise an exception indicating 
that value does not find.</li>
        </ul>
    <li>Insertion</li>
        <ul>
            <li>Syntax : list.insert(index,obj) </li>
            <li>Description : The method insert() inserts object obj into list at offset index.</li>
            <li>Parameters : seq -- This is the list of elements</li>
                <ul>
                    <li>index -- This is the Index where the object obj need to be inserted.</li>
                    <li>obj -- This is the Object to be inserted into the given list.</li>
                </ul>
            <li>Return Value : This method does not return any value but it inserts the given element at the given 
index.</li>
        </ul>
    <li>Pop/Deletion</li>
        <ul>
            <li>Syntax : list.pop(obj=list[-1])</li>
            <li>Description : The method pop() removes and returns last object or obj from the list.</li>
            <li>Parameters : obj -- This is an optional parameter, index of the object to be removed from 
the list.</li>
            <li>Return value : This method returns the removed object from the list. </li>
        </ul>
    <li>Remove</li>
        <ul>
            <li>Syntax : List.remove(obj) </li>
            <li>Description : Removes an object from the list.</li>
            <li>Parameters : obj -- This is the object to be removed from the list. </li>
            <li>Return value : This method does not return any value but removes the given object from the list.  </li>
        </ul>
    <li>Reverse</li>
        <ul>
            <li>Syntax : List.reverse() </li>
            <li>Description : The method reverse() reverses objects of list in place.</li>
            <li>Parameters : NA </li>
            <li>Return value : This method does not return any value but reverse the given object from the list.</li>
        </ul>
    <li>Sorting</li>
        <ul>
            <li>Syntax : list.sort([func])</li>
            <li>Description : Use the sort() method to sort a list in ascending order.</li>
            <li>Parameters : </li>
                <ul>
                    <li>key (optional): A function that takes one argument and returns a value to use for sorting</li>
                    <li>reverse (optional): Boolean. If True, sorts in descending order. Default is False.</li>
                </ul>
            <li>Return value : This method does not return any value but reverse the given object from the list.</li>
        </ul>
</ol>
<ul>
    <li>Use the key parameter to define custom sorting criteria. Pass a function that specifies the sorting rule. Example: my_list.sort(key=len) (Sorts by the length of elements).</li>
    <li>For case-insensitive sorting, use str.lower as the key.<br> Example: my_list.sort(key=str.lower)</li>
    <li>To keep the original list unchanged, use the sorted() function instead of sort(). It returns a new sorted list.<br>Example: new_list = sorted(my_list)</li>
    <li>To sort in descending order, set the reverse parameter to True. Example: my_list.sort(reverse=True)</li>
    <li>Combine key and reverse for advanced sorting, such as sorting by custom rules in descending order.<br>
    Example: my_list.sort(key=lambda x: x[-1], reverse=True) (Sorts by the last character in reverse order).</li>
</ul>

```python
#List comprehension
squares = [x**2 for x in range(10)]
print("Squares of numbers from 0 to 9:", squares)
# List Methods

# 1. append() method
numbers = [1, 2, 3, 4, 5]
numbers.append(6)
print("List after appending 6:", numbers)

#  2. extend() method
more_numbers = [7, 8, 9]
numbers.extend(more_numbers)
print("List after extending with more_numbers:", numbers)

# 3. insert() method
numbers.insert(0, 0)
print("List after inserting 0 at the beginning:", numbers)

# 4. remove() method
numbers.remove(3)
print("List after removing 3:", numbers)

# 5. pop() method
popped_element = numbers.pop()
print("Popped element:", popped_element)

# 6. index() method
index_of_4 = numbers.index(4)
print("Index of 4:", index_of_4)

# 7. count() method
count_of_2 = numbers.count(2)
print("Count of 2 in the list:", count_of_2)

# 8. sort() method
numbers.sort()
print("List after sorting:", numbers)

# 9. reverse() method
numbers.reverse()
print("List after reversing:", numbers)

# 10. clear() method
numbers.clear()
print("List after clearing:", numbers)

# 11. copy() method
numbers_copy = numbers.copy()
print("Copy of the list:", numbers_copy)

# Sorting using parameters
numbers = [5, 2, 9, 1, 5, 6]

# Sorting in ascending order
numbers.sort()
print("List after sorting in ascending order:", numbers)

# Sorting in descending order
numbers.sort(reverse=True)  
print("List after sorting in descending order:", numbers)
# Sorting using a custom key function
def custom_key(x):
    return x % 3

numbers.sort(key=custom_key)
print("List after sorting with custom key:", numbers)

# Sorting a list of strings based on their lengths
words = ["apple", "banana", "kiwi", "grape", "orange"]
words.sort(key=len)
print("List of words sorted by length:", words)

# Sorting a list of tuples based on the second element
tuples_list = [(1, 'one'), (3, 'three'), (2, 'two'), (4, 'four')]
tuples_list.sort(key=lambda x: x[1])
print("List of tuples sorted by the second element:", tuples_list)
```

```python
# Functions on Lists

# 12. len() function
length_of_numbers = len(numbers)
print("Length of the list:", length_of_numbers)

# 13. sum() function
sum_of_numbers = sum(numbers)
print("Sum of the list:", sum_of_numbers)

# 14. min() function
min_of_numbers = min(numbers)
print("Minimum value in the list:", min_of_numbers)

# 15. max() function
max_of_numbers = max(numbers)
print("Maximum value in the list:", max_of_numbers)

# 16. any() function
any_true = any(numbers)
print("Any element in the list is True:", any_true)

# 17. all() function
all_true = all(numbers)
print("All elements in the list are True:", all_true)

# 18. enumerate() function
for index, value in enumerate(numbers):
    print(f"Index: {index}, Value: {value}")

# 19. filter() function
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print("Even numbers in the list:", even_numbers)

# 20. map() function
squared_numbers = list(map(lambda x: x**2, numbers))
print("Squared numbers in the list:", squared_numbers)
```
